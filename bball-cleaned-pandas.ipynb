```python
import pandas as pd
import numpy as np
import json
```

```python
data = json.load(open('bball-scraped.json'))
```

```python
data[0]
```

```python
data_for_df = []
```

```python
#loop through, get stats if they exist 
for player in data:
    for key, val in player.get('stats', {}).get('career_stats', {}).items():
        stat_type = key.lower().split()[0]
        name = player.get('Name')

        for stat in val:
            stat['name'] = name
            stat['stat_type'] = stats_type
            data_for_df.append(stat)
```

```python
df = pd.DataFrame(data_for_df)
```
