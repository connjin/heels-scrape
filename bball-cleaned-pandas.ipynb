```python
import pandas as pd
import numpy as np
import json
```

```python
data = json.load(open('bball-scraped.json'))
```

```python
data[0]
```

```python
data_for_df = []
```

```python
for player in data:
    for key, val in player.get('stats', {}).get('current_stats', {}).items():
        stats_type = key.lower().split()[0]
        name = player.get('Name')
      
        for stat in val:
            stat['name'] = name
            stat['stat_type'] = stats_type
            data_for_df.append(stat)
```

```python
#dataframe!!!!!!! it's lit
df = pd.DataFrame(data_for_df)
```

```python
df
```

```python
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.drop('avg', axis=1, inplace=True)
```

```python
#make groups
# by_year = df.groupby('year'); 
by_player = df.groupby(('name', 'stat_type'))
```
